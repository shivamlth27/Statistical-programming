# Q1.
 A publishing firm wants to develop special printing machines for English. For this, they need to
determine the probability of occurrence of specific letters and words. You are given two large text
files (file A and file B).

1. Determine the probability of each alphabet in the English language. Uppercase and lower-
case alphabets are considered the same. Other characters like whitespace, punctuation, and
numerals are to be omitted. List the top ten occurring alphabets. Is the result the same for
file A and file B?

```python
import string
import math


# defining function 'clean_string' for removing whitespace, punctuation, and numerals and
# changing into lower case
def clean_string(s):
    s=s.lower()
    s=s.replace(" ","")
    s=s.replace(".","")
    s=s.replace("\n","")
    return  s
```

function for calculating probability of each alphabet which return dictionary 'prob_alpha'

```python
def prob_cal(file_path):

    # opening the file and reading file content
    with open(file_path, 'r') as file:
        file_content = file.read()

    #clearing the file content by using 'clean_string' function
    input_string=clean_string(file_content)

    #making a list 'count' for storing count of each alphabet
    count=[0]*26
    a_z=string.ascii_lowercase

    #counting the number of each alphabet
    a=-1
    for j in a_z:
        a+=1
        for i in input_string:
            if(j==i):
                count[a]+=1
    
    #calculating probability of each alphabet
    prob=[]
    for i in range(0,26):
        prob.append(count[i]/sum(count))

    #adding probability of each alphabet into dictionary
    a_z=string.ascii_lowercase
    prob_alpha={}
    for i in range(0,26):
        prob_alpha[a_z[i]]=prob[i]
    return prob_alpha
```

## Entropy Calculation
2. The measure of uncertainty is determined by its entropy. Entropy should be computed as
<p align='center'>H = ∑(−pi log pi)</p>
where pi is the probability of event i. If we consider the occurrence of alphabets in English
as events of interest, determine the entropy. In other words, determines the uncertainty of
the alphabet in the English language. Use the results from the previous question.

```python
# defining function 'entropy_cal' for calculating entropy
def entropy_cal(file_path):
    req_dic=prob_cal(file_path)
    entropy=0
    for key in req_dic:
        entropy+=-req_dic[key] *math.log2(req_dic[key])
    return entropy
```

```python
l=list(range(16))
r_state= [f'r{i}' for i in l]
rv_R={}
for i in range(0,16):
    rv_R[r_state[i]]=r_norm[i]
rv_R
```

```python
cdf=[0]*16
for i in range(16):
    for j in range(i+1):
        cdf[i]+=r_norm[j]
cdf
```

```python
round_S=[]
for i in rv_S:
    round_S.append(round(i))
round_S
```

function to generate uniform random number y

```python
import time

def custom_uniform():
    # Use the current time in seconds as a seed
    seed = int(time.time() * 1000000)
    
    # Define constants for the linear congruential generator
    a = 1664525
    c = 1013904223
    m = 2**32

    # Linear congruential generator
    seed = (a * seed + c) % m
    
    # Map the seed to a number between 0 and 1
    random_number = seed / m

    return random_number
```

## 2.
Utilize the grayscale values of the chosen image from the previous problem. Apply a random
variable transformation to the gray level values to obtain a uniform PMF.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('moon.jpg', cv2.IMREAD_GRAYSCALE)

# Flatten the image array to a 1D array
flat_image = image.flatten()

# Compute the histogram and CDF of the original image
hist, bins = np.histogram(flat_image, bins=256, range=[0, 256], density=True)
cdf = hist.cumsum()

# Transform the image using the CDF to obtain a uniform PMF
transformed_image = np.interp(flat_image, bins[:-1], cdf * 255).reshape(image.shape).astype(np.uint8)

# Display the original and transformed images
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.imshow(image, cmap='gray')
ax1.set_title('Original Image')

ax2.imshow(transformed_image, cmap='gray')
ax2.set_title('Transformed Image (Uniform PMF)')

plt.show()
```
